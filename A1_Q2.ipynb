```python
import numpy as np
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python

dataset = load_dataset("fashion_mnist")
X_train = np.array([np.array(img).flatten() for img in dataset['train']['image']]) / 255.0
y_train = np.array(dataset['train']['label'])
X_test = np.array([np.array(img).flatten() for img in dataset['test']['image']]) / 255.0
y_test = np.array(dataset['test']['label'])
```

```python

X_train = np.hstack((np.ones((X_train.shape[0], 1)), X_train))
X_test = np.hstack((np.ones((X_test.shape[0], 1)), X_test))
```

```python

def softmax(z):
    z -= np.max(z, axis=1, keepdims=True) # Numerical stability
    exp_z = np.exp(z)
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

def one_hot(y, num_classes):
    return np.eye(num_classes)[y]

def compute_gradient(X, y_true, y_pred):
    m = X.shape[0]
    return (1/m) * np.dot(X.T, (y_pred - y_true))

def gradient_descent(X, y, lr=0.1, epochs=100):
    num_classes = len(np.unique(y))
    weights = np.zeros((X.shape[1], num_classes))
    y_hot = one_hot(y, num_classes)

    for epoch in range(epochs):
        scores = np.dot(X, weights)
        y_pred = softmax(scores)


        grad = compute_gradient(X, y_hot, y_pred)
        weights -= lr * grad

        if epoch % 10 == 0:
            loss = -np.mean(np.sum(y_hot * np.log(y_pred + 1e-8), axis=1))
            print(f"Epoch {epoch}, Loss: {loss:.4f}")

    return weights
```

```python
weights = gradient_descent(X_train, y_train, lr=0.5, epochs=100)
```

```python
test_scores = np.dot(X_test, weights)
test_preds = np.argmax(softmax(test_scores), axis=1)
accuracy = np.mean(test_preds == y_test)
print(f"FashionMNIST Test Accuracy: {accuracy * 100:.2f}%")
```
